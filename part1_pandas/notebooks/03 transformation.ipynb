```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/financials_clean.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['profit_margin'] = round((df['net_profit']/df['sales'])*100,2)
```

```python
df[['company_name','financial_year','profit_margin']].head()
```

```python
df.head()
```

```python
df['opm']=round((df['operating_profit']/df['sales'])*100,2)
```

```python
df.head()
```

```python
df['expense_ratio']=round(df['expenses']/df['sales']*100,2)
```

```python
df.head()
```

```python
df['interest_coverage'] = round(
    df['operating_profit'] / df['interest'],
    2
)
```

```python
df.head()
```

```python
df['interest_burden_ratio'] = round(
    (df['interest'] / df['sales']) * 100,
    2
)
```

```python
df['tax_burden_ratio'] = round(
    (df['net_profit'] / df['profit_before_tax']) * 100,
    2
)
```

```python
df.head()
```

```python
df = df.sort_values(['company_name','financial_year'])
```

```python
df['revenue_growth']=round(df.groupby('company_name')['sales'].pct_change(fill_method=None)*100,2)
```

```python
df[
    [
        'company_name',
        'financial_year',
        'sales',
        'revenue_growth'
    ]
].head(15)
```

```python
df['profit_growth'] = round(df.groupby('company_name')['net_profit'].pct_change(fill_method = None)*100,2)
```

```python
df.head(15)
```

```python
cagr_df = df.groupby('company_name').agg({'financial_year':['min','max'],'sales':['first','last']})
```

```python
cagr_df.head()
```

```python
cagr_df.columns = [
    'start_year',
    'end_year',
    'beginning_sales',
    'ending_sales'
]
```

```python
cagr_df.head()
```

```python
cagr_df['years'] = (
    cagr_df['end_year']
    - cagr_df['start_year']
)
```

```python
cagr_df['cagr']=round(((cagr_df['ending_sales']/cagr_df['beginning_sales'])**(1/cagr_df['years'])-1)*100 ,4)
```

```python
cagr_df.head()
```

```python
df = df.merge(
    cagr_df[['cagr']],
    on='company_name',
    how='left'
)
```
