```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/financials.csv")
```

```python
df.shape
```

```python
df = df.rename(columns={"Unnamed: 0": "metric"})
```

```python
df.columns[:5]
```

```python
year_columns = [
    'Mar 2012','Mar 2013','Mar 2014',
    'Mar 2015', 'Mar 2016', 'Mar 2017',
'Mar 2018', 'Mar 2019', 'Mar 2020',
'Mar 2021', 'Mar 2022', 'Mar 2023']
```

```python
cols_to_keep = ['metric','company_name'] + year_columns
```

```python
df_filtered = df[cols_to_keep]
```

```python
df_filtered.head(11)
```

```python
df_filtered.shape
```

```python
df_long = pd.melt(df_filtered,id_vars=['metric','company_name'],var_name='financial_year',value_name='value')
```

```python
df_long.head()
```

```python
df_long.shape
```

```python
df_long['financial_year'] = (
    df_long['financial_year']
    .str.replace('Mar ', '')
    .astype(int)
)
```

```python
df_long.head()
```

```python
df_long.dtypes
```

```python
df_long.head(80000)
```

```python
important_metrics = [
    'Sales',
    'Expenses',
    'Operating Profit',
    'Net Profit',
    'EPS in Rs',
    'Interest',
    'Depreciation',
    'Profit before tax'
]
```

```python
df_long = df_long[df_long['metric'].isin(important_metrics)]
```

```python
df_long = df_long.dropna(subset=['value'])
```

```python
df_long.dropna(subset=['value'])
```

```python
df_pivot = df_long.pivot_table(index=['company_name','financial_year'],
                               columns='metric',
                               values='value').reset_index()
```

```python
df_pivot.head()
```

```python
df_pivot.columns.name = None
```

```python
df_pivot.columns = (df_pivot.columns.str.lower().str.replace(' ','_'))
```

```python
df_pivot.head()
```

```python
df_pivot = df_pivot[
    [
        'company_name',
        'financial_year',
        'sales',
        'expenses',
        'operating_profit',
        'interest',
        'depreciation',
        'profit_before_tax',
        'net_profit',
        'eps_in_rs'
    ]
]
```

```python
df_pivot.to_csv(
    '../data/processed/financials_clean.csv',
    index=False
)
```
